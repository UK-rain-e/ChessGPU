```python
from time import sleep
from IPython.display import display, clear_output
import os
from time import time

import chess
import chess.pgn
import chess.engine


from openai import OpenAI
```

```python
client = OpenAI()
```

```python
pgn = open("game.pgn")
game = chess.pgn.read_game(pgn)
board = game.board()
engine = chess.engine.SimpleEngine.popen_uci("/opt/homebrew/bin/stockfish")    
```

```python
messages = [
    {"role": "system", 
     "content": """
        You are a chess commentator.
        Don't mention you are a language model, just comment the game in a fun and entertaining manner.
        Each user message will be either a chess move, or an absence of the move (player is still thinking). 
        It will also contain full game position and relative evaluation in centi-pawns.
        You need to comment the move, or state of the game as a whole. Suggest potential further developments, potential moves and threats.
        Don't include move number into your response. Also don't directly say stockfish eval value, it's for your information only.
        Feel free to add historical references to famous chess games and jokes. 
        Generate no more than 5 sentences for each response.
        After you comment a move, wait for user to give you a new move, you don't make any moves yourself.
    """
    }
]

skip_first = 0
move_idx = 0

for move in game.mainline_moves():    
    move_idx += 1
#     clear_output(wait=True)
#     print(move)
    board.push(move)
    display(board)
    info = engine.analyse(board, chess.engine.Limit(time=0.3))
    print("Score:", info["score"])
    
    message = {
        "role": "user", "content": f"Move: {str(move)}. Stockfish eval: {str(info['score'])}"
    }
    messages.append(message)
    
    
    if move_idx <= skip_first:
        continue
        
    start = time()
        
    response = client.chat.completions.create(
      model="gpt-4-1106-preview",
      messages=messages
    )
    
    finish = time()
    print(f"Duration: {finish-start}")
    
    response = response.choices[0].message.content
    print(response)
    
    message = {
        "role": "assistant", "content": response
    }
    messages.append(message)
    
#     os.system(f'./talk.sh "{response}" &> /dev/null');


#     sleep(5)
```

```python

board = chess.Board()
while not board.is_game_over():
    clear_output(wait=True)
    result = engine.play(board, chess.engine.Limit(time=0.5))
    board.push(result.move)
    display(board)

engine.quit()
```
